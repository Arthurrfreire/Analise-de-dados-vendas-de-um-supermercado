```python
import pandas as pd

# Carregar o conjunto de treinamento
train_data = pd.read_csv('Train-Set.csv')

# Vizualizar as primeiras linhas do conjunto de treinamento
print(train_data.head())

```

```python
# Resumo estatístico das variáveis numéricas
print(train_data.describe())

# Contagem de valores únicos nas variáveis categóricas
print(train_data.select_dtypes(include='object').nunique())
```

```python
import matplotlib.pyplot as plt

# Calcular a média das vendas em cada loja
sales_by_outlet = train_data.groupby('OutletID')['OutletSales'].mean().sort_values()

# Criar gráfico de barras
plt.figure(figsize=(10, 6))
sales_by_outlet.plot(kind='bar', color='skyblue')
plt.title('Média de Vendas por Loja')
plt.xlabel('Vendas Médias')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Gráfico de dispersão entre MRP e vendas
plt.figure(figsize=(10, 6))
plt.scatter(train_data['MRP'], train_data['OutletSales'], alpha=0.5, color='orange')
plt.title('Relação entre MRP e Vendas')
plt.xlabel('MRP (Preço Máximo de Varejo)')
plt.ylabel('Vendas')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calcular as vendas médias por tipo de produto
sales_by_product_type = train_data.groupby('ProductType')['OutletSales'].mean().sort_values()

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
sales_by_product_type.plot(kind='bar', color='lightgreen')
plt.title('Vendas Médias por Tipo de Produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Vendas Médias')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calcular as vendas médias por tamanho da loja
sales_by_outlet_size = train_data.groupby('OutletSize')['OutletSales'].mean().sort_values()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sales_by_outlet_size.plot(kind='bar', color='lightblue')
plt.title('Vendas Médias por Tamanho da Loja')
plt.xlabel('Tamanho da Loja')
plt.ylabel('Vendas Médias')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calcular as vendas médias por tipo de localização da loja
sales_by_location_type = train_data.groupby('LocationType')['OutletSales'].mean().sort_values()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sales_by_location_type.plot(kind='bar', color='lightcoral')
plt.title('Vendas Médias por Tipo de Localização da Loja')
plt.xlabel('Tipo de Localização da Loja')
plt.ylabel('Vendas Médias')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calcular as vendas médias por ano de estabelecimento da loja
sales_by_establishment_year = train_data.groupby('EstablishmentYear')['OutletSales'].mean().sort_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sales_by_establishment_year.plot(kind='bar', color='lightpink')
plt.title('Vendas Médias por Ano de Estabelecimento da Loja')
plt.xlabel('Ano de Estabelecimento da Loja')
plt.ylabel('Vendas Médias')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Criar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(train_data['ProductVisibility'], train_data['OutletSales'], alpha=0.5, color='purple')
plt.title('Relação entre Visibilidade do Produto e Vendas')
plt.xlabel('Visibilidade do Produto')
plt.ylabel('Vendas')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
